```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x**5 + 4 * np.sin(2*x) + np.cos(3*x + 3))
```

```python
print(f"f'(x) = {derivative(f, 1.0, dx=1e-6)},\nf''(x) = {derivative(f, 1.0, dx=1e-6, n = 2)}")
```

```python
X = np.linspace(-5, 5)
f1 = [derivative(f, x, dx = 1e-6) for x in X]
f2 = [derivative(f, x, dx = 1e-6, n=2) for x in X]
plt.plot(X, f1, label="f'(x)")
plt.plot(X, f2, label="f''(x)")
plt.legend();
```
