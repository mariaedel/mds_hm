```python
import numpy as np
import matplotlib.pyplot as plt
import math

```

```python
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])

u = x + y + z
```

```python
plt.plot(x[0], x[1], 'o', label='x')
plt.plot(y[0], y[1], 'o', label='y')
plt.plot(z[0], z[1], 'o', label='z')

plt.plot(u[0], u[1], 'o', label='x + y + z')

plt.grid()
plt.legend();
```

```python
# угол между x и y в градусах

fi = math.acos(np.dot(x, y) / ((x[0]**2 + x[1]**2)**(1/2) + (y[0]**2 + y[1]**2)**(1/2))) * 57.2958
fi
```
