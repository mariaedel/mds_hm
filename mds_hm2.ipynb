## Уменьшить в 2 раза

```python
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
A = A / 2
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

## Повернуть на 130 градусов (относительно oX)

```python
fi = 130*0.0174533
# U = np.array([[1, 0, 0], [0, np.cos(fi), -np.sin(fi)], [0, np.sin(fi), np.cos(fi)]])
U = np.array([[np.cos(fi), -np.sin(fi), 0], [np.sin(fi), np.cos(fi), 0], [0, 0, 1]])

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
A = A @ U
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

## Отразить относитльно y = x

```python
U = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 0]])

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
A = A @ U
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

## Сместить на (200, 300) и масштабировать на (0.5, 1.2)

```python
S = np.array([[1, 0, 0], [0, 1, 0], [200, 300, 1]]) # матрица переноса
M = np.array([[0.5, 0, 0], [0, 1.2, 0], [0, 0, 1]]) # матрица масштабирования

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
A = A @ S @ M
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([100, 500])
plt.xlim([0, 300])
```
